```python
'''
This notebook creates the model comparison figure
'''
# import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    "Model": ["Transformer", "Base ESM", "30-Layered ESM", "ESM-SXGbg Fold", "Transformer", "Base ESM", "30-Layered ESM", "ESM-SXGbg Fold", "Transformer", "Base ESM", "30-Layered ESM", "ESM-SXGbg Fold", "Transformer", "Base ESM", "30-Layered ESM", "ESM-SXGbg Fold", "SXGbg Fold", "SXGbg Fold", "SXGbg Fold", "SXGbg Fold"],
    "Dataset": ["DD", "DD", "DD", "DD", "EDD", "EDD", "EDD", "EDD", "TG", "TG", "TG", "TG", "SCOPe", "SCOPe", "SCOPe", "SCOPe", "DD", "EDD", "TG", "SCOPe"],
    "Accuracy": [0.871, 0.829, 0.876, 0.931, 0.953, 0.903, 0.951, 0.964, 0.874, 0.843, 0.924, 0.933, 0.822, 0.655, 0.800, 0.922, 0.859, 0.958, 0.888, 0.837]
} # results recorded as seen (inefficient, but it works)

df = pd.DataFrame(data) # create a pandas dataframe
df.head() # show the first 5 rows of the dataframe
```

```python
sns.barplot(x="Dataset", y="Accuracy", hue="Model", data=df, palette="viridis") # create the barplot
plt.xlabel("Dataset", size=14) # assign x label
plt.ylabel("Accuracy", size=14) # assign y label
plt.legend(loc = "lower center") # assign legend location
fig = plt.gcf() # get current figure
fig.savefig("results.png", dpi=300, bbox_inches="tight") # save the figure
plt.show() # show the figure
```
